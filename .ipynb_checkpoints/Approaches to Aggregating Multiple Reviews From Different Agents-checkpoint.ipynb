<h1>Approaches to Aggregating Multiple Reviews From Different Agents</h1>

<p>We will explore 7 approaches to aggregating reviews from multiple agents</p>

<ul>
  <li><p>Let us assume we have two agents <b>x</b> and <b>y</b></p></li>
  <li>There are total set of arguments proposed by agents is <b>[a,b,c]</b></li>
  <li>Arguments <b>[a,c]</b> are postive and <b>b</b> is negative</li>
  <li>Agent <b>x</b> proposes arguments <b>[a,b,c]</b> and votes positive</li>
  <li>Agent <b>y</b> proposes arguments <b>[a,b]</b> and votes positive</li>
</ul>




<h2>Approach 1: No enrichment</h2>
In this approach we say that an agent can only distribute mass into argument graphs that contain exactly the arguments proposed by the agent.

So:
<ul>
<li>Agent <b>x</b> can only distribute mass into argument containing [a,b,c]</li>
<li>Agent <b>y</b> can only distribute mass into argument containing [a,b]</li>
</ul>

<h2>Approach 2: Can enrich graphs that have more nodes than you</h2>
In this approach we say that an agent can distribute mass into argument graphs that contain exactly the arguments proposed by the agent <b>and</b> to graphs that contain more nodes than it.

An way to think about this would be to say had an agent had access to those additional arguments then he would have distributed mass into those graphs that matches theirs. 

So:
<ul>
<li>Agent x can only distribute mass into argument containing [a,b,c]</li>
<li>Agent y can only distribute mass into argument containing [a,b] <b>and</b> [a,b,c]</li>
</ul>

<h2>Approach 3: Can enrich graphs that have less nodes than you</h2>
In this approach we say that an agent can distribute mass into argument graphs that contain exactly the arguments proposed by the agent <b>and</b> to graphs that contain less nodes than it.

An way to think about this would be to say that although an agent has proposed an argument we do not know how much they would like to have the argument in the argument graph. Hence the agent would possibly like to distribute mass to graphs with fewer arguments.

So:
<ul>
<li>Agent x can only distribute mass into argument containing [a,b,c] <b>and</b> [a,b] <b>and</b> [a,c] <b>and</b> [a] <b>and</b> [b] <b>and</b> [c]</li>
<li>Agent y can only distribute mass into argument containing [a,b]</li>
</ul>

<h2>Approach 4: Can enrich graphs that have more nodes than you and less nodes</h2>
In this approach we say that an agent can distribute mass into argument graphs that contain exactly the arguments proposed by the agent <b>and</b> to graphs that contain more or less nodes than it.

This allows agents to adopt both approach 2 and 3

So:
<ul>
<li>Agent x can only distribute mass into argument containing [a,b,c] and [a,b] and [a,c] and [a] and [b] and [c]</li>
<li>Agent y can only distribute mass into argument containing [a,b] and [a,b,c]</li>
</ul>

<h2>Approach 5: Can enrich graphs that have more nodes than you but contributions are to larger graphs are weighted</h2>
In this approach we say that an agent can distribute mass into argument graphs that contain exactly the arguments proposed by the agent <b>and</b> to graphs that contain more nodes than it.

The difference from Approach 2 is that we say an agent proposes arguments and so this is suggests to an extent that they would those arguments in the argument graph. However larger graphs should also be contributed to but not as much.


So:
<ul>
<li>Agent x can only distribute mass into argument containing [a,b,c]</li>
<li>Agent y can only distribute mass into argument containing [a,b] and [a,b,c]</li>
</ul>

We would say here that agent y here believes in graphs containing [a,b] twice as much as graphs containing [a,b,c]. 

The addition of an extra node means that subset of gaphs are halving their chances to recieve mass

So we say that [a,b] is twice as likely as [a,b,c] i.e 

<p>[a,b]:[a,b,c] :: 2 : 1 </p>

That means graphs with [a,b] have 2/3 of the weight and graphs with [a,b,c] have 1/3

<p>[a,b,c]:[a,b]:[a,c]:[a]:[c] :: 4:2:2:1:1 </p>
